```python
from cgitb import text
import unittest
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import os
from webdriver_manager.chrome import ChromeDriverManager
from tqdm import tqdm
import pandas as pd
import bangla


os.environ["PATH"] += r"F:\selenium driver"  #Driver Path

driver = webdriver.Chrome(ChromeDriverManager().install())  # Installing Drivers


bible_para_xpath = "/html/body/div[2]/div/div/div[6]/div/p"  #xpath of the location of the bible verses
chapter_title_xpath = "/html/body/div[2]/div/div/div[1]/h1"  #xpath of the location of chapter title
chapter_xpath = "/html/body/div[2]/div/div/div[4]/div/div/div[2]/div[2]/ul/li/a" #xpath of the location of chapter title

bible_para_lst = [] #here all the verses will be stored 
chapter_title = [] # here all the chapter title will be stored
chapter_name = []  # here all the chapter name will be stored
chapter_no = []  #chapter numbers will be stored
verse_no = []  #verses numbers will be stored


for k in tqdm(range(1,40,1)):   #as the last book no of old testament is 39
    flag = True
    k = str(k)
    if len(k) == 1:
        k = "0"+k  #to match with website url
    for l in tqdm(range(1,151)):  #highest amount of chapter a book can have is 150

        if flag == False:  
            break

        url = f"https://www.wordproject.org/bibles/ben/{k}/{l}.htm#0" #setting up the url of the page
        driver.get(url)
        driver.implicitly_wait(5)  #waiting for the website to load
        try:
            yoyo = driver.find_element(By.XPATH,"/html/body/center[1]/h1")  #if the page shows 404
            if yoyo.text == "404 Not Found":  #if it shows 404 we will go to the next book
                flag = False
                break
        except: #if it doesn't shows 404 we will do this






            elem = driver.find_elements(By.XPATH,bible_para_xpath) #it contains whole pages verses
            #here elem is a tuple

            word = ""
            
            chapter_title_web = driver.find_element(By.XPATH,chapter_title_xpath)#it contains the title of the chapter
            #print((elem))
            for elm in elem:  #it only runs 1 as elem only have one elm in it 

                


                result = ""

                tempnum = 1
                for i in elm.text.split(): #making elm as text 
                    
                    if i.isdigit(): #condition when to count the numbers when not to 
                        temp_i = int(i)-1 

                        if temp_i == tempnum:  #this condition applies when we will not count the number
                            
                        
                            result += "GG" +" "  #as a tag for splitting 
                            tempnum = int(i)
                        else:
                            result +=bangla.convert_english_digit_to_bangla_digit(i) + " " #this condition applies when we will count number
                    
                    else:
                        result += i + " "
                fresult = result.split("GG")
                
                result = fresult #it contains verses separately

                
                    
                count = 1
                
                for i in result: #for each verses

                    chapter_title.append(chapter_title_web.text) #chapter name only 


                    chpt_name_temp = f"{chapter_title_web.text} {str(bangla.convert_english_digit_to_bangla_digit(l))}:{bangla.convert_english_digit_to_bangla_digit(str(count))}"
                    chapter_name.append(chpt_name_temp)  #chapter name big
                    chapter_no.append(bangla.convert_english_digit_to_bangla_digit(l)) #chapter number
                    verse_no.append(bangla.convert_english_digit_to_bangla_digit(count)) #verse number
                    
                    
                    
                    bible_para_lst.append(i) #verse 
                    count += 1
```

```python
print(len(chapter_name))
print(len(chapter_title))
print(len(chapter_no))
print(len(verse_no))
print(len(bible_para_lst))

data = {
    "citation":chapter_name,
    "book":chapter_title,
    "chapter":chapter_no,
    "verse": verse_no,
    "text":bible_para_lst
}
df = pd.DataFrame(data)
print(df)
df.to_csv (r'F:\Bible\export_dataframe.csv', index = False, header=True)
```
